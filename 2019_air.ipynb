#### Dropping column 58 which is a long string with all the values of this DF.

```python
df.drop('CODI_PROVINCIA;PROVINCIA;CODI_MUNICIPI;MUNICIPI;ESTACIO;CODI_CONTAMINANT;ANY;MES;DIA;H01;V01;H02;V02;H03;V03;H04;V04;H05;V05;H06;V06;H07;V07;H08;V08;H09;V09;H10;V10;H11;V11;H12;V12;H13;V13;H14;V14;H15;V15;H16;V16;H17;V17;H18;V18;H19;V19;H20;V20;H21;V21;H22;V22;H23;V23;H24;V24', axis='columns', inplace=True)
```

```python
df = df.select_dtypes(exclude = object).copy()
```

```python
df
```

```python
df.drop(columns=['CODI_PROVINCIA', 'CODI_MUNICIPI'], inplace=True)
```

```python
df
```

#### Creating Date column converting years, months and days from floats to strings and putting together to get a date.

```python
df['DATA'] = (df["ANY"].astype(str).apply(lambda x: x[:-2]) + "/"  + df["MES"].astype(str).apply(lambda x: x[:-2]) + "/" + df["DIA"].astype(str).apply(lambda x: x[:-2]))
```

```python
df
```

```python
df['DATA'] = pd.to_datetime(df['DATA'], dayfirst=False, errors='coerce')
```

```python
df
```

```python
df.drop(columns=['ANY', 'MES', 'DIA'], inplace=True)
```

```python
df
```

```python
df['ESTACIO'] = df['ESTACIO'].astype(str).apply(lambda x: x[:-2])
df['CODI_CONTAMINANT'] = df['CODI_CONTAMINANT'].astype(str).apply(lambda x: x[:-2])
```
