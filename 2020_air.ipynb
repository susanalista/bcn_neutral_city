# Air Measurements from April to October 2020

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../bcn_open_data/air_quality_2020.csv',index_col=[0])
```

```python
df = df.select_dtypes(exclude = object).copy()
```

```python
df
```

```python
df.drop(columns=['CODI_PROVINCIA', 'CODI_MUNICIPI'], inplace=True)
```

```python
df
```

#### Creating Date column joining years, months and days from integers to strings and putting together to get a date.

```python
df['DATA'] = (df["ANY"].astype(str).apply(lambda x: x) + "/"  + df["MES"].astype(str).apply(lambda x: x) + "/" + df["DIA"].astype(str).apply(lambda x: x))
```

```python
df
```

```python
df['DATA'] = pd.to_datetime(df['DATA'], dayfirst=False, errors='coerce')
```

```python
df
```

#### Dropping unnecessary columns for year, month and day

```python
df.drop(columns=['ANY', 'MES', 'DIA'], inplace=True)
```

#### Converting integers for station and polluter ID to strings

```python
df['ESTACIO'] = df['ESTACIO'].astype(str)
df['CODI_CONTAMINANT'] = df['CODI_CONTAMINANT'].astype(str)
```

#### Filling NaN with the previous column value

```python
df.bfill(axis=0)
```

#### Saving clean dataset to csv

```python
df.to_csv('air_measurements_2020.csv', index=False)
```
