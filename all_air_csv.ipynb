## Splitting "Air Measurements" csv by year

```python
import pandas as pd
import os
```

```python
file_list = [f for f in os.listdir('../bcn_open_data/Aire/2019') if f.endswith('.csv')]
print(file_list)
```

```python
data_sets = [pd.read_csv(os.path.join('../bcn_open_data/Aire/2019', f)) for f in file_list]
data = pd.concat(data_sets, axis=0)
data.head()
```

```python
data.to_csv('air_quality_2019.csv') 
```

```python
file_list1 = [f for f in os.listdir('../bcn_open_data/Aire/2020') if f.endswith('.csv')]
print(file_list1)
```

```python
data_sets1 = [pd.read_csv(os.path.join('../bcn_open_data/Aire/2020', f)) for f in file_list1]
data1 = pd.concat(data_sets1, axis=0)
data1.head()
```
